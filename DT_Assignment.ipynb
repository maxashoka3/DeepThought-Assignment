### Processing for output file

```python
# checking both dataset  columns 
input_file_1.columns, input_file_2.columns
```

```python
# merging the both input file
merged_df = pd.merge(input_file_1, input_file_2,left_on = 'User ID',right_on = 'uid')
```

```python
#viewing
merged_df
```

```python
# grouping the team name by total_statements and by total_reasons and rouding the values
new_df = merged_df.groupby('Team Name')['total_statements','total_reasons'].mean().round(2).reset_index()            
```

```python
# sorting the according to the total_statetements
new_df.sort_values('total_statements',ascending = False,inplace = True)
```

```python
# renaming the columns to the approprate name
new_df = new_df.rename(columns ={
    'Team Name': 'Thinking Teams Leaderboard',
    'total_statements':'Average Statements',
    'total_reasons':'Average Reasons'}).reset_index(drop = True)
```

```python
# Adding the rank
new_df.reset_index(inplace = True)
new_df['Team Rank'] = new_df.index +1
```

```python
# viewing 
new_df
```

```python
# deleting the index columns
new_df.drop('index',axis = 1,inplace = True)
```

```python
# viewing the columns
new_df.columns
```

```python
# selecting the required columns
new_df = new_df[['Team Rank','Thinking Teams Leaderboard', 'Average Statements', 'Average Reasons']]
new_df
```

```python

# viewing the new dataset
merged_df
```

```python
# checking the columns of merged dataset
merged_df.columns
```

```python
# selecting the required colummns for processing
leader_indi =merged_df[['Name','uid','total_statements','total_reasons']]
```

```python
#Sorting by total_statements and by total_reasons
leader_indi = leader_indi.sort_values(['total_statements','total_reasons'],ascending=[False,False])
```

```python
# Adding the rank 
leader_indi.reset_index(inplace = True)
Rank = leader_indi.index +1
```

```python
#creating the new columns
leader_indi['Rank'] = Rank
```

```python
# renaming the columns in the proper format
leader_indi.rename(columns = {
    'uid': 'UID',
    'total_statements':'No. of Statements',
    'total_reasons':'No. of Reasons'
    
},inplace = True)
```

```python
#checking the columns
leader_indi.columns
```

```python
# combining the required columns
leader_indi = leader_indi[['Rank', 'Name','UID','No. of Statements','No. of Reasons']]
leader_indi
```
